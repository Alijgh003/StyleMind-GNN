```python
import pandas as pd
```

```python
categories_path = "categories.csv"
```

```python
categories_df = pd.read_csv(categories_path)
categories_df = categories_df.rename(columns={'undefined': 'fine_grained', 'Unnamed: 2': 'category', '2':"ID"})
categories_df
```

```python
categories_df['category'] = categories_df['category'].apply(lambda x: x.strip() if isinstance(x, str) else x)
categories_df
```

```python
categories_df[categories_df['category'].isna()]
##We have checked the items metadata and understood that no item has belonged to
## following categories
```

```python
import math
catToID = {}
c = 1
for i in categories_df['category']:
  if(isinstance(i,float) and math.isnan(i)):
    catToID[i] = c
    c+=1
  else:
    j = i.strip()
    if(j not in catToID):
      catToID[j] = c
      c+=1
catToID
```

```python
categories_df['category_ID'] = categories_df['category'].apply(lambda x: catToID[x])
categories_df
```

```python
map_IDs = {i:j for i,j in zip(categories_df['ID'], categories_df['category_ID'])}
map_IDs
```

```python
items_df = pd.read_csv("items.csv")
items_df
```

```python
items_df['category_id'] = items_df['category_id'].apply(lambda x: map_IDs[x])
items_df
```

```python
items_df.to_csv("new_items.csv")
categories_df.to_csv("old_categories.csv")
```

```python
final_categories = pd.DataFrame(catToID.items(), columns=['category', 'category_id'])
final_categories.to_csv("new_categories.csv")
final_categories
```

```python
datasets_path = path.join(data_path,"dataset","polyvore_outfits","nondisjoint")
```

```python
import json
def read_dataset_json(split , datasets_path):
  f = open(path.join(datasets_path,split+".json"), "r")
  dataset_json = json.loads(f.read())
  flatten_dataset = [{**item, 'set_id': outfit['set_id']} for outfit in dataset_json for item in outfit['items']]
  f.close()
  return flatten_dataset
train_set = read_dataset_json("train", datasets_path)
test_set = read_dataset_json("test", datasets_path)
valid_set = read_dataset_json("valid", datasets_path)
train_set
```

```python
def edge_indices_df(flatten_dataset, split, datasets_path):
  dataset_df = pd.DataFrame(flatten_dataset)
  dataset_df['item_id'] = dataset_df['item_id'].apply(lambda x: int(x))
  dataset_df['semantic_category'] = dataset_df['item_id'].apply(lambda x: items_df.loc[x]['semantic_category'])
  dataset_df['item_index'] = dataset_df['item_id'].apply(lambda x: items_df.loc[x]['index'])
  df_pairs = dataset_df.merge(dataset_df, on='set_id')
  df_pairs = df_pairs[df_pairs['item_id_x'] < df_pairs['item_id_y']]
  df_pairs.to_csv(path.join(datasets_path,split+"_edges.csv"))
  return df_pairs

train_edges_df = edge_indices_df(train_set, "train",datasets_path)
test_edges_df = edge_indices_df(test_set, "test", datasets_path)
valid_edges_df = edge_indices_df(valid_set, "valid", datasets_path)
valid_edges_df
```

```python
def get_maximum_number_of_edge_in_outfit(edges_df):
  maximum_number_of_edges = edges_df.groupby("set_id")['index_y'].count().max()
  return maximum_number_of_edges

print(f'train: {get_maximum_number_of_edge_in_outfit(train_edges_df)}')
print(f'test: {get_maximum_number_of_edge_in_outfit(test_edges_df)}')
print(f'valid: {get_maximum_number_of_edge_in_outfit(valid_edges_df)}')
```

```python
import torch
def create_edge_index_tensor(edges_df,split,datasets_path=None, save=True):
  maximum_number_of_edges = get_maximum_number_of_edge_in_outfit(edges_df)
  edge_index_by_setid_df = edges_df.groupby("set_id")[['item_index_x','item_index_y']].apply(lambda x: list(zip(x['item_index_x'],x['item_index_y'])))
  edge_index_list = [[*i, *[(-1,-1)]*(maximum_number_of_edges - len(i))] for i in edge_index_by_setid_df]
  edge_index_tensor = torch.tensor(edge_index_list).mT
  if(save):
    torch.save(edge_index_tensor,path.join(datasets_path,split+"_edge_index.pth"))
  return edge_index_tensor

train_edge_index_tensor = create_edge_index_tensor(train_edges_df,"train",datasets_path, True)
valid_edge_index_tensor = create_edge_index_tensor(valid_edges_df,"valid",datasets_path, True)
test_edge_index_tensor = create_edge_index_tensor(test_edges_df,"test",datasets_path, True)
print(f'train: {train_edge_index_tensor.shape}')
print(f'valid: {valid_edge_index_tensor.shape}')
print(f'test: {test_edge_index_tensor.shape}')
```

```python
def select_edge_indices(edge_index_tensor_by_groups):
  result = torch.cat([edge_index_tensor_by_groups[i] for i in range(edge_index_tensor_by_groups.shape[0])],dim=-1)
  result = result[:,torch.where(result[0]>-1)[0]]
  return result

sampled_edge_indices = train_edge_index_tensor[torch.randperm(train_edge_index_tensor.shape[0])]
edges = select_edge_indices(sampled_edge_indices)
edges.shape
```

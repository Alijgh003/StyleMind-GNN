```python
file = open("./polyvore/train_no_dup.json","r")
train_data = file.read()
file.close()
len(train_data)
```

```python
import json

json_train_data = json.loads(train_data)
def get_garment_ID(garment):
    return garment['image'].split("=")[-1]
garment = json_train_data[4]['items'][0]
searchingID = get_garment_ID(garment)
searchingID, garment
```

```python
total_items = 0
total_items = [garment for outfit in json_train_data for garment in outfit['items']]
distinct_items = {get_garment_ID(garment) for garment in total_items}
print(len(total_items) / len(json_train_data))
len(distinct_items),len(total_items)
```

```python
# reading category list
file = open("./polyvore/category_id.txt","r")
categories = file.read().split("\n")
categories = {int(ca.split(maxsplit=1)[0]):ca.split(maxsplit=1)[1] for ca in categories}
file.close()
len(categories.keys()),categories
```

```python
relations_dict = {}
for outfit in json_train_data:
    for g1 in outfit['items']:
        for g2 in outfit['items']:
            if(get_garment_ID(g1) != get_garment_ID(g2)):
                r1 = g1['categoryid'] if g1['categoryid']< g2['categoryid'] else g2['categoryid']
                r2 = g1['categoryid'] if g1['categoryid']>= g2['categoryid'] else g2['categoryid']
                relations_dict[(r1,r2)] = relations_dict.get((r1,r2),1)

print(f'the number of relations: {len(relations_dict)}/{190*379}')
relations_dict
```

```python
l = list(relations_dict)
for i,v in relations_dict.items():
    if v>1:
        print(i,v)
        break
```

```python
# map categories
import ast
file = open("./polyvore/map_categories.txt", "r")
mcats = file.read().split('\n')

mcats = {ast.literal_eval(m[:-1])[0]: ast.literal_eval(m[:-1])[1] for m in mcats}
file.close()
mkeys = {value for _, value in mcats.items()}
len(mkeys),mkeys, len(mcats.keys())
```

```python
mmap = {}
errors = []
for id, category in categories.items():
    try:
        mmap[id] = mcats[category]
    except KeyError:
        errors.append(category)
file = open("./errors.txt","w")
file.write(str(errors))
file.close()
mmap
```
